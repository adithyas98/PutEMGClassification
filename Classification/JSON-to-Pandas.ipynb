```python
import json
import os
# import pandas as pd
import numpy as np
# df=pd.DataFrame()
count=0
dataset=[]
for filename in os.listdir(data_path)[:100:10]:
  with open(data_path+filename) as f:
    data = json.load(f)
    # df=df.append(pd.json_normalize(data))
    li=[]
    for i in list(data['data'].keys()):
      npdata=np.array(data['data'][i])
      li.append(np.pad(npdata,(0,256000-len(npdata)),'constant'))  #some windows smaller than 256000
    ar=np.array(li)
    dataset.append(ar)
  count+=1
  if(count%10==0):
    print(count)
dataset1=np.array(dataset)
```
